```python
# https://www.youtube.com/watch?v=H05fMVpl8Lo&list=PLxgDUj5eygKnSd4MYHIuAC7E3VEPm_WLO&index=13
# LINE CHART
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create sample data
x = np.linspace(0, 20, 1_000)
# evenly spaced points on x-axis b/w 0 and 20 divided into 1,000 equally-spaced sections
```

```python
# calculate some ys for the xs just generated
y = np.sin(x) + 0.2 * x
```

```python
# plot data
plt.plot(x, y)
```

```python
# in a notebook, this won't do anything yet
# --> call additional SHOW function
plt.plot(x, y)
plt.show()
```

```python
# add some titles
plt.plot(x, y)
plt.xlabel("input")
plt.ylabel("output")
plt.title("some fancy function with sin")
```

```python
# notebooks always print the output of the last statement, in this case:
# Text(0.5, 1.0, 'some fancy function with sin')
# to suppress that, END THE LAST STATEMENT WITH A SEMICOLON (;)
plt.plot(x, y)
plt.xlabel("input")
plt.ylabel("output")
plt.title("some fancy function with sin");
```
