```python
# https://www.youtube.com/watch?v=-CMhzgFbEtg&list=PLxgDUj5eygKnSd4MYHIuAC7E3VEPm_WLO&index=14
# SCATTERPLOT
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 100 observations x 2 dimensions (for x and y coordinates)
X = np.random.randn(100, 2)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# classification --> 2 classes
X = np.random.randn(200, 2)
# move 1/2 half to be centred at 3
X[:50] += 3
# generate labels
Y = np.zeros(200)
# first half to label 1
Y[:50] = 1
# draw scatterplot
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.show()
```
