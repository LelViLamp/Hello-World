```python
from PIL import Image
```

```python
im = Image.open('downloaded_image.png')
# does not return a np array
type(im)
```

```python
# convert into np array
# works b/c internally represented as array (array-like structure)
arr = np.array(im)
```

```python
arr
```

```python
arr.shape
# 258 pixels high
# 249 pixels wide
# 3 for the three colours (RGB)
```

```python
plt.imshow(arr)
plt.show()
```

```python
# works w/ original image too
plt.imshow(im)
plt.show()
```

```python
grey = arr.mean(axis=2)
grey.shape
```

```python
plt.imshow(grey)
plt.show()
```

```python
# weird colours... but I want them in black and white!!!!
# decided in background by matplotlib --> HEATMAP
plt.imshow(grey, cmap='gray')
plt.show()
```
